```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
```

```python
digits = load_digits()
```

```python
X = digits.data
y = digits.target
```

```python
plt.figure(figsize=(10, 4))
for index, (image, label) in enumerate(zip(digits.images[:8], digits.target[:8])):
    plt.subplot(1, 8, index + 1)
    plt.imshow(image, cmap='gray')
    plt.title(f'Digit: {label}')
    plt.axis('off')
plt.show()
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=64)
X_pca = pca.fit_transform(X_scaled)
```

```python
explained_variance = np.cumsum(pca.explained_variance_ratio_)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(explained_variance, marker='o', linestyle='--')
plt.title('Cumulative Explained Variance by Principal Components')
plt.xlabel('Number of Principal Components')
plt.ylabel('Cumulative Explained Variance')
plt.grid(True)
plt.show()
```

```python
pca_2d = PCA(n_components=2)
X_pca_2d = pca_2d.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(10, 6))
scatter = plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], c=y, cmap='viridis', s=30)
plt.colorbar(scatter)
plt.title('PCA of Digits Dataset (2D)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
print(f"Explained variance by first 2 components: {pca_2d.explained_variance_ratio_}")
```

```python
pca_3d = PCA(n_components=3)
X_pca_3d = pca_3d.fit_transform(X_scaled)
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca_3d[:, 0], X_pca_3d[:, 1], X_pca_3d[:, 2], c=y, cmap='viridis', s=30)
plt.colorbar(scatter)
ax.set_title('PCA of Digits Dataset (3D)')
ax.set_xlabel('PC 1')
ax.set_ylabel('PC 2')
ax.set_zlabel('PC 3')
plt.show()
```
